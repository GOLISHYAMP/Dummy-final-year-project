# Machine Learning Algorithm

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
x=covid.drop('COVID-19',axis=1)
y=covid['COVID-19']
```

```python
x.columns
```

```python
y.name
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.8, random_state = 88, shuffle = True, stratify = y )
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
LR = LogisticRegression()
LR.fit(x_train,y_train)
y_pred = LR.predict(x_test)
acc_LR = accuracy_score(y_test,y_pred)*100
print(acc_LR)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(n_estimators=1000)
RFR.fit(x_train, y_train)
y_pred = RFR.predict(x_test)
acc_RFR= RFR.score(x_test, y_test)*100
print(acc_RFR)
#acc_RFR = roc_auc_score(y_test, y_pred)*100
#print(acc_RFR)
```

## KNeighborClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
# finding the best number for KNeighbors
accuracy_train =[]
accuracy_test = []
for i in range(1,51):
  KNC = KNeighborsClassifier(n_neighbors = i)
  KNC.fit(x_train,y_train)
  accuracy_test.append(KNC.score(x_test,y_test))
  accuracy_train.append(KNC.score(x_train,y_train))
plt.plot(range(1,51),accuracy_test,label = 'test')
plt.plot(range(1,51),accuracy_train,label = 'train')
plt.legend()
plt.show()  
```

## From the above graph we come to know that best Kneighbor number will be 12

```python
KNN = KNeighborsClassifier(n_neighbors=12)
KNN.fit(x_train, y_train)
y_pred = KNN.predict(x_test)
#acc_KNN=KNN.score(x_test, y_test)*100
#print(acc_KNN)
#acc2_KNN = roc_auc_score(y_test,y_pred)*100
#print(acc2_KNN)
acc_KNN = accuracy_score(y_test,y_pred)*100
print(acc_KNN)
```

## DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier()
DTC.fit(x_train,y_train)
y_pred = DTC.predict(x_test)
#acc_DTC = DTC.score(x_test,y_test)*100
#print(acc_DTC)
#acc2_DTC = roc_auc_score(y_test,y_pred)*100
#print(acc2_DTC)
acc_DTC = accuracy_score(y_test,y_pred)*100
print(acc_DTC)
```

## naive_bayes

```python
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
NB.fit(x_train,y_train)
y_pred = NB.predict(x_test)
acc_NB = accuracy_score(y_test,y_pred)*100
print(acc_NB)
```

# SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
parameters = {'kernel':['linear','rbf','sigmoid','poly'], 'gamma':[1,0.1,0.01]}
grid = GridSearchCV(SVC(), parameters, verbose=2, scoring= 'neg_mean_squared_error')
grid.fit(x,y)
best_params = grid.best_params_
print(best_params)
```

```python
SVMC = SVC(kernel='sigmoid',gamma=0.01)
SVMC.fit(x_train, y_train)
y_pred = SVMC.predict(x_test)
acc_SVMC= accuracy_score(y_test,y_pred)*100
print(acc_SVMC)
```

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes',   
              'Decision Tree'],
    'Score': [acc_SVMC, acc_KNN, acc_LR, 
              acc_RFR, acc_NB, acc_DTC]})
models.sort_values(by='Score', ascending=False)
```
